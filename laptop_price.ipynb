```python
companies = ['Apple', 'Dell', 'Lenovo', 'HP', 'Asus', 'Acer', 'MSI', 'toshiba']

def convert_to_company(x):
    if x not in companies:
        return 'Others'
    return x
```

```python
df['Company'] = df['Company'].apply(convert_to_company)
```

```python
df.info()
```

```python
df['TypeName'].value_counts()
```

```python
df['ScreenResolution'].value_counts()
```

```python
cpus = df['Cpu'].value_counts()
```

```python
def convert_cpus(x):
    if cpus[x] > 1:
        return x
    return 'others'
```

```python
df['Cpu'] = df['Cpu'].apply(convert_cpus)
```

```python
df.info()
```

```python
df['Ram'] = df['Ram'].apply(lambda x: int(x[:-2]))
```

```python
def process_memory(mem):
    mem = mem.replace('GB', '').replace('TB', '000')
    ssd = 0
    hdd = 0
    flash = 0
    hybrid = 0

    parts = mem.split('+')

    for part in parts:
        part = part.strip()
        if 'SSD' in part:
            ssd += float(part.split()[0])
        elif 'HDD' in part:
            hdd += float(part.split()[0])
        elif 'Flash Storage' in part:
            flash += float(part.split()[0])
        elif 'Hybrid' in part:
            hybrid += float(part.split()[0])
    return pd.Series([ssd, hdd, flash, hybrid])
```

```python
df[['SSD', 'HDD', 'Flash', 'Hybrid']] = df['Memory'].apply(process_memory).astype(int)
```

```python
df.info()
```

```python
df.drop('Memory', axis = 1, inplace = True)
```

```python
df['Gpu'] = df['Gpu'].apply(lambda x: x.split()[0])
```

```python
df['OpSys'].unique()
```

```python
df['Weight'] = df['Weight'].apply(lambda x: float(x[:-2]))
```

```python
df.info()
```

```python
df['Price_rupees'] = df['Price_euros']*105
```

```python
df = df.drop('Price_euros', axis = 1)
```

```python
df.head()
```

```python
X = df.drop('Price_rupees', axis = 1)
Y = df['Price_rupees']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
trf = ColumnTransformer(transformers = [
    ('ohe', OneHotEncoder(handle_unknown = 'ignore'), [0, 1, 3, 4, 6, 7]),
    ('scaling', StandardScaler(), [2, 5, 8, 9, 10, 11, 12])
], remainder = 'passthrough')
```

```python
pipe = make_pipeline(trf, RandomForestRegressor())
```

```python
pipe.fit(X, Y)
```

```python
ypred = pipe.predict(X_test)
```

```python
r2_score(Y_test, ypred)
```

```python
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(pipe, f)
```
